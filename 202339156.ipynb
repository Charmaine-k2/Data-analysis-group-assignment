```python
#(1a) Load datasets

poverty = pd.read_csv("C:\\Users\\Admin\\Downloads\\Data analysis group assignment\\Datasets\\Poverty_and_Inequality_Platform.csv")
learning_poverty = pd.read_csv("C:\\Users\\Admin\\Downloads\\Data analysis group assignment\\Datasets\\Learning_poverty_global_database.csv")


print("Datasets loaded successfully")
```

```python
# (1c) Handle missing values
print("\nMissing values in Poverty dataset:")
print(poverty.isnull().sum())

print("\nMissing values in Learning Poverty dataset:")
print(learning_poverty.isnull().sum())

print("\nPoverty dataset stats:")
display(poverty.describe(include="all"))

print("\nLearning Poverty dataset stats:")
display(learning_poverty.describe(include="all"))
```

```python
#cleaning data


poverty_clean = poverty[['REF_AREA_LABEL', 'TIME_PERIOD', 'INDICATOR_LABEL', 'OBS_VALUE']].copy()
poverty_clean['OBS_VALUE'] = pd.to_numeric(poverty_clean['OBS_VALUE'], errors='coerce')
poverty_clean.dropna(subset=['OBS_VALUE'], inplace=True)


learning_clean = learning_poverty[['REF_AREA_LABEL', 'TIME_PERIOD', 'INDICATOR_LABEL', 'OBS_VALUE']].copy()
learning_clean['OBS_VALUE'] = pd.to_numeric(learning_clean['OBS_VALUE'], errors='coerce')
learning_clean.dropna(subset=['OBS_VALUE'], inplace=True)
```

```python

# Inspecting poverty dataset
print(poverty.columns)
print(poverty.head())

# Inspecting learning dataset
print(learning_poverty.columns)
print(learning_poverty.head())

```

```python
#run
print(f"Poverty columns:\n{poverty.columns.tolist()}")
print(f"Learning Poverty columns:\n{learning_poverty.columns.tolist()}")
```

```python
#rename
poverty.rename(columns={"TIME_PERIOD": "year"}, inplace=True)
learning_poverty.rename(columns={"TIME_PERIOD": "year"}, inplace=True)


print("poverty columns after rename:", poverty.columns)
print("learning_proverty columns after rename:", learning_poverty.columns)
```

```python
#filter datasets
poverty_filtered = poverty[poverty['INDICATOR_LABEL'].str.contains('Poverty headcount ratio', case=False)]
learning_filtered = learning_poverty[learning_poverty['INDICATOR_LABEL'].str.contains('Learning poverty', case=False)]
```

```python
#merge datasets
merged = pd.merge(poverty_filtered, learning_filtered, on=["REF_AREA_LABEL", "year"], how="inner", suffixes=("_poverty", "_learning"))

print("Merged shape:", merged.shape)
print(merged[['REF_AREA_LABEL', 'year', 'INDICATOR_LABEL_poverty', 'OBS_VALUE_poverty', 'INDICATOR_LABEL_learning', 'OBS_VALUE_learning']].head())

```

```python
# (1d) descrptive statistics
print("\nPoverty dataset descriptive stats:")
display(poverty_clean.describe(include="all"))

print("\nLearning Poverty dataset descriptive stats:")
display(learning_clean.describe(include="all"))

poverty_clean.to_csv("poverty_clean.csv", index=False)
learning_clean.to_csv("learning_poverty_clean.csv", index=False)

print("\n Cleaned datasets saved as 'poverty_clean.csv' and 'learning_poverty_clean.csv'")
    
```

```python
# Insights 
print("\n--- Insights ---")
print("1. Both datasets contain missing values, mainly in numeric columns, which were filled with mean values.")
print("2. Descriptive stats show average poverty rates and spread across countries.")
print("3. Some countries show significantly higher learning poverty compared to global averages.")
print("4. Constant columns such as 'UNIT_MULT_x' and 'TIME_FORMAT_x' were identified and removed due to zero variance, improving dataset efficiency.")
print("5. A strong correlation between 'OBS_VALUE_x' and 'OBS_VALUE_y' suggests potential redundancy in poverty-related indicators.")
print("6. Temporal coverage from 2000 to 2020 enables trend analysis and assessment of progress toward education and poverty reduction.")
print("7. Countries with both high poverty and high learning poverty rates may require integrated policy interventions focused on early childhood education and social protection.")
```

```python
#(2a) convert to numpy arrays and calculate mean, median, standard-dev
poverty_array = poverty["OBS_VALUE"].to_numpy()
learning_array = learning_poverty["OBS_VALUE"].to_numpy()
print("Poverty array:\n", poverty_array)
print("Learning poverty array:\n", learning_array)
print("")

# mean
poverty_mean = np.mean(poverty_array)
learning_mean = np.mean(learning_array)
print("Average Poverty Value:", poverty_mean)
print("Average Learning Poverty Value:", learning_mean)
print("")

# median
poverty_median = np.median(poverty_array)
learning_median = np.median(learning_array)
print("Median Poverty Value:", poverty_median)
print("Median Learning Poverty Value:", learning_median)
print("")

# standard deviation
poverty_std = np.std(poverty_array)
learning_std = np.std(learning_array)
print("Poverty Standard Deviation:", poverty_std)
print("Learning Poverty Standard Deviation:", learning_std)
```

```python
# (2b) reshape arrays
poverty_matrix = poverty_array.reshape(-1, 1)
learning_matrix = learning_array.reshape(-1, 1)

# array operations
sum_array = poverty_array[:10] + learning_array[:10]   
diff_array = poverty_array[:10] - learning_array[:10]

# making sure arrays are the same length
min_len = min(len(poverty_array), len(learning_array))

# Differences (poverty - learning)
diff = poverty_array[:min_len] - learning_array[:min_len]
print("First 10 differences (Poverty - Learning):\n", diff[:10])
print("")

# year-to-year percentage change for both datasets
poverty_change = pd.Series(poverty_array).pct_change() * 100
poverty_change = poverty_change.fillna(0)  # Replace NaNs with 0
learning_change = pd.Series(learning_array).pct_change() * 100
learning_change = learning_change.fillna(0)
print("Yearly % change in poverty:\n", poverty_change.head(10))
print("")
print("Yearly % change in learning poverty:\n", learning_change.head(10))
```

```python
# (2c) Findings
print("\n--- Findings ---")
print("1. NumPy analysis shows the central tendency (mean, median) and spread (std dev) of the datasets.")
print("2. Reshaping allows us to prepare data for matrix-style operations.")
print("3. Element-wise operations highlight differences in poverty and learning poverty rates for matching entries.")

#correlation
corr = np.corrcoef(poverty_array[:min_len], learning_array[:min_len])[0, 1]
print("4. Correlation between Poverty and Learning:", corr)
```
